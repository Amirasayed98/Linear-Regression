## 1. Let's Now Import library of pandas and numpy




```python
import pandas as pd
import numpy as np
```

## 2. Load The Data into a Pandas Frame , hint  use **pd.read_csv**

```python
df = pd.read_csv('kc_house_data.csv')
df.head()
```

```python
df.info()
```

```python
sum(df.duplicated())
```

### 3.Drop (id,date) column from x(independent variables ),  hint use **drop()** function

```python
df_cleaned =df.drop(['id','date'], 1)
df_cleaned.head()
```

```python
sum(df_cleaned.duplicated())
```

```python
df_cleaned.drop_duplicates(inplace=True)
```

## 4. define x is the features of dataframe and y is label , hint use **df.iloc**

```python
X = df_cleaned.iloc[:,1:] 
X.head()
```

```python
Y = df_cleaned.iloc[:, :1]
Y.head()
```

## 5. Splitting the dataset into the Training set and Test set and put **random_state=2**


```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
Y_train.head()
```

```python
Y_test.head()
```

## 6. Fitting Linear Regression to the dataset and count regression.score

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
```

```python
regressor.score(X_train, Y_train)
```

```python
regressor.score(X_test, Y_test)
```
